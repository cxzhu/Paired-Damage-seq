```python
import snapatac2 as snap
import anndata
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import plotly.io as pio

os.chdir("/gpfs/commons/groups/zhu_lab/zcao/Paired_damage/data/SnapATAC2_whitelist/")
```

```python
whitelist_df = pd.read_csv("/gpfs/commons/groups/zhu_lab/zcao/Paired_damage/data/RNA/metadata_011024_seu.csv")
whitelist_df
```

```python
whitelist = whitelist_df["barcodes"].to_list()
```

```python
#setting treatment groups
treatments = ["Ctrl", "0hr", "2hr", "6hr", "24hr", "48hr"]
```

```python
for treatment in treatments:
    file_path = "/gpfs/commons/groups/zhu_lab/zcao/Paired_damage/data/fragments/SnapATAC2_{}_fragment.tsv.gz".format(treatment)
    file_name = "snapATAC2_10000_{}.h5ad".format(treatment)
    data = snap.pp.import_data(file_path, chrom_sizes=snap.genome.hg38, 
                               sorted_by_barcode=False, file=file_name, n_jobs = 16, whitelist = whitelist)
    snap.pp.add_tile_matrix(data, bin_size = 10000) #5000/10000
    snap.pp.select_features(data, n_features=25000)
    snap.tl.spectral(data)
    snap.tl.umap(data)
    snap.pp.knn(data)
    snap.tl.leiden(data, resolution = 0.5)
    snap.pl.umap(data, color='leiden', interactive=False, height=500, 
                 out_file = "/gpfs/commons/groups/zhu_lab/zcao/Paired_damage/data/SnapATAC2_whitelist/Plots/UMAP_10000_{}.pdf".format(treatment))
    data.close()
```
